# 添加这个模块来学习pandas的使用以及画图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data from txt
data_1 = pd.read_csv('data.txt',header=None,names=['x','y'])
# 将其放置到坐标系上
data_1.plot(kind='scatter',x='x',y='y',figsize=(5,5))
plt.show()
```

```python
# 画直线等
fig ,ax = plt.subplots()
# 取1000个采样点
x = np.linspace(data_1.x.min(),data_1.x.max(),1000)
# -3.24140214  1.1272942, 设出欲画出的函数
f_x = -3.24140214 + 1.1272942 * x
# 添加函数
ax.plot(x,f_x,'green',label='f(x)')
ax.plot(x,10+np.sqrt(10-np.power(x-12.5,2)),'black',label='g(x)')

# 加上散点图
ax.scatter(data_1.x,data_1.y,label='Scatter Point')
ax.scatter([11,22,13],[12,13,16],label='Test')

# 设置标签
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Test')



# 图例的位置
ax.legend(loc=4)

#展示
plt.show()
```
